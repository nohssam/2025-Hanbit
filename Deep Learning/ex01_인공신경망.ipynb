### 로지스틱 회귀로 패션 아이템 분류하기

```python
# 이미지의 픽셀값을 0~1 범위로 정규화(normalization) 하는 과정
train_scaled = train_input / 255.0
# 로지스틱 회귀는 3차원을 사용할 수 없음 (2차원배열)
# 2차원으로 변경 (60000*28*28) => (60000*(28*28))
train_scaled = train_scaled.reshape(-1, 28*28)
```

```python
# 3차원(60000,28,28) => 2차원(60000,(28*28)) (사이키런의 모델에 사용가능)
print(train_scaled.shape)
```

```python
# 교차 검증
from sklearn.model_selection import cross_validate
# 경사하강법
from sklearn.linear_model import SGDClassifier

sc = SGDClassifier(loss='log_loss', max_iter=5, random_state=42)
scores = cross_validate(sc, train_scaled, train_target, n_jobs=-1)
print(np.mean(scores['test_score']))
```

```python
# 10번 반복
sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42)
scores = cross_validate(sc, train_scaled, train_target, n_jobs=-1)
print(np.mean(scores['test_score']))
```

```python
sc = SGDClassifier(loss='log_loss', max_iter=20, random_state=42)
scores = cross_validate(sc, train_scaled, train_target, n_jobs=-1)
print(np.mean(scores['test_score']))
```

### 인공신경망으로 모델 만들기

```python
# 텐서플로 : 구글이 2015년 11월 오픈 소스로 공개한 딥러닝 라이브러리 
import tensorflow as tf 
# 텐서플로에는 저수준 API와 고수준API가 있다.
# 케라스가 텐서플로의 고수준 API이다.(모델을 정의하는 API만 있고 실제 모델의 계산을 수행하는 부분은 텐서에 위임한다.
# 케라스는 인공 신경망(딥러닝) 모델을 “쉽고 빠르게” 만들 수 있게 도와주는 고수준(High-level) API
import keras 
import os 
os.environ["KERAS_BACKEND"] = "torch" # 또는 "jax"
```

```python
# 현재 Keras가 어느 백엔드를 사용 중인지 알려주는 함수
keras.config.backend() 
```

```python
from sklearn.model_selection import train_test_split

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
print(train_scaled.shape, train_target.shape)
```

```python
print(val_scaled.shape, val_target.shape)
```

```python
# 케라스에서 입력층을 정의할 때는 다음처럼 Input()함수(클래스아이다)를 사용
# Input 층 정의
# 매개변수에 입력의 크기를 지정(여기에서는 784개의 픽셀이 1창원으로 펼쳐져있음)
inputs = keras.layers.Input(shape=(784,))
```

```python
# 밀집층
# 뉴런의 개수는 10개, 
# activation='softmax' => 뉴런에 출력에 적용할 함수
# 활성화 함수로는 softmax 사용(다중분류=> 확률로 변경)
# 만약에 이진분류이면 시그모이드 함수 사용 (activation='sigmoid')
dense = keras.layers.Dense(10, activation='softmax')
display(Image(filename="data/img05.png", width=600, height=400))
```

```python
# 케라스의 Sequential 클래스를 이용해서 입력층과 밀집층을 가진 신경망 모델을 만들
model = keras.Sequential([inputs, dense])
display(Image(filename="data/img06.png", width=600, height=400))
```

### 인공 신경망으로 패션 아이템 분류하기 

```python
# 손실함수 지정, : sparse_categorical_crossentropy
# 케라스에서는  이진분류일때는  binary_crossentropy
#               다중분류일때는 categorical_crossentorpy
# 아래 그림 처럼 0,1을 주는 것이 아니라 정수값을 그대로 사용한다는 의미에서 sparse (스파스) 사
# 측정지표 지정 : 정확도
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
display(Image(filename="data/img07.png", width=600, height=400))
display(Image(filename="data/img08.png", width=600, height=400))
```

```python
print(train_target[:10])
```

```python
# 훈련
model.fit(train_scaled, train_target, epochs=5)
```

```python
# 평가 점수
# 손실값 : 0.4444445073604584
# 정확도 : 0.8458333611488342
model.evaluate(val_scaled, val_target)
```
