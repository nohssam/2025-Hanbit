### 결정트리
>- 데이터의 특징을 기준으로 조건분기(if-then규칙)를 만들어 분류 또는 회귀 문제를 푸는 알고리즘
>- 사이킷 런의 DecisionTreeClassifier 클래스를 사용해서 결정 트리 모델를 훈련할 수 있다.

```python
import pandas as pd
# 데이터 가져오기
wine = pd.read_csv('https://bit.ly/wine_csv_data')
data = wine[wine.columns[:-1]]
target = wine[wine.columns[-1]]

# 훈련세트와 테스트세트 나누기
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state = 42)

# 표준화 처리 
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
## 분류용 결정 트리 모델
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_scaled, train_target)
print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target)) # 과적합되어있음
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
# 모델이 학습한 모든 분기 조건을 그림으로 보여줌
plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()
```

```python
# 모델이 학습한 모든 분기 조건을 그림으로 보여줌
plt.figure(figsize=(10,7))
# max_depth=1 -> 첫번째 분기 기준만 보여줌
plot_tree(dt, max_depth=1, filled=True, feature_names=train_input.columns)
plt.show()
```

```python
from IPython.display import Image, display
display(Image(filename="data/img21.png", width=600, height=400))
# 불순도란 한 노드 안에 서로 다른 클래스가 얼마나 섞여 있는가를 나타내는 지표이다.
# 불순도(Gini)가 0 -> 오직 한 클래스만 존재(완전히 순수함)
# 불순도(Gini)가 0.5 -> 두 클래스가 반반으로 섞여 있음
# 불순도 값이 작을수록 -> 한쪽 글래스가 훨씬 많음 -> 분류가 명확
# 불순도 값이 클수록 -> 클래스가 섞여 있음  -> 분류가 불명확
# 결정 트리가 데이터를 순수한 그룹으로 점점 나누는  과정 이라고 생각하면 된다.
# 왼쪽으로 갈수록 조건을 만족하는 데이터들이 모이는 방향이다.
# 왼쪽이 오른쪽보다 항상 순수하다고는 할 수 없다.
```

### 가지치기
>- 결정트리가 너무 복잡하게 자라는 것을 막기 위해, 불필요한 가지를 잘라내는 과정
>- 즉 트리가 너무 깊어지면 훈련 데이터에 과도하게 맞춰지는 과적합이 발생한다.
>- 이름 방지하기 위해, 트리의 일부 가지를 잘라내거난 성장을 제한하다.
>- 이 과정을 **가지치기** 라고 부른다.

```python
# max_depth = 3  => 가지치기
dt = DecisionTreeClassifier(max_depth=3,random_state=42)
dt.fit(train_scaled, train_target)
print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target)) # 과적합되어있음
```

```python
# 각 노드는 가장 많은 비중을 차지한 클래스의 색깔로 채워진다.
# 파란색 -> 클래스 0 
# 주황색 -> 클래스 1
# 색의 진한 색은 불순도가 낮음

plt.figure(figsize=(10,7))
# max_depth=1 -> 첫번째 분기 기준만 보여줌
plot_tree(dt, filled=True, feature_names=train_input.columns)
plt.show()
```

```python
print(dt.feature_importances_)
# 결정 트리가 학습 과정에서 각 특성이 결정(분류)에 얼마나 기여했는지를 나타내는 **중요도 점수**이다.
# alcohol (12.3%), suger(86.8%), pH(0.7%)
```
