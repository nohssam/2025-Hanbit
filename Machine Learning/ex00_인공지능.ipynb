```python
인공지능이란
사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템을 만드는 기술

인공지능이란
인공 일반 지능(Artificial general Intelligence) 혹은 강 인공 지능(Strong AI)
영화에 흔히 등장하는 인공지능
영화 <그녀>의 사만다나 지금도 가장 사악한 인공지능으로 불리는 <터미네이터>의 스카이넷처럼 사람과 구분하기 어려운 지능을 가진 컴퓨터 시스템
약인공지능(Week AI)
현실에서 우리가 마주하고 있는 인공지능
아직까지는 특정 분야에서 사람의 일을 도와주는 보조 역할만 가능(음성 비서, 자율 주행 자동차, 음악 추천, 기계 번역 등)
예: 이세돌과 바둑 시합을 한 알파고
```

```python
머신러닝이란
규칙을 일일이 프로그래밍하지 않아도 자동으로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야
인공지능의 하위 분야 중에서 지능을 구현하기 위한 소프트웨어를 담당하는 핵심 분야
머신러닝은 통계학과 깊은 관련이 있음
통계학에서 유래된 머신러닝 알고리즘이 많으며, 통계학과 컴퓨터 과학 분야가 상호 작용하면서 발전
대표적인 오픈소스 통계 소프트웨어인 R에는 다양한 머신러닝 알고리즘이 구현되어 있음
최근 머신러닝의 발전은 통계나 수학 이론보다 경험을 바탕으로 발전하는 경우도 많음
컴퓨터 과학 분야가 이런 발전을 주도, 컴퓨터 과학 분야의 대표적인 머신러닝 라이브러리는 사이킷런
사이킷런이 있기 전까지 머신러닝 기술은 대부분 폐쇄적인 코드와 라이브러리로 통용되어서 해당 분야에 대한 전문 교육을 이수하거나 비싼 비용을 지불하고 구매를 해야 했음
하지만 사이킷런과 같은 오픈소스 라이브러리의 발전 덕분에 머신러닝 분야는 말 그대로 폭발적으로 성장
파이썬 코드를 다룰 수 있다면 누구나 머신러닝 알고리즘을 무료로 손쉽게 제품에 활용할 수 있음
```

```python
딥러닝이란
딥러닝(deep learning): 많은 머신러닝 알고리즘 중에 인공 신경망(artificial neural network)을 기반으로 한 방법들의 통칭
대부분 인공 신경망과 딥러닝을 크게 구분하지 않고 사용
최초의 합성곱 신경망: 1998년 얀 르쿤(Yann Lecun)이 신경망 모델 LeNet-5을 만들어 손글씨 숫자를 인식하는 데 성공
합성곱 신경망이 이미지 분류 작업에 널리 사용되기 시작: 2012년에 제프리 힌턴(Geoffrey Hinton)의 팀이 AlexNet 모델로 이미지 분류 대회인 ImageNet에서 기존의 머신러닝 방법을 누르고 우승
국내에서는 2016년 이세돌과 알파고의 대국으로 인해 딥러닝에 대한 관심이 크게 상승
2022년에는 챗GPT가 출시되면서 일반 대중이 인공지능 기술을 직접 활용하는 시대를 열었습니다.
```

```python
인공지능은 사람처럼 학습하고 추론할 수 있는 지능을 가진 시스템을 만드는 기술
인공지능은 강인공지능과 약인공지능으로 구분
머신러닝은 규칙을 프로그래밍하지 않아도 자동으로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야
사이킷런이 대표적인 라이브러리
딥러닝은 인공 신경망이라고도 하며, 텐서플로와 파이토치가 대표적인 라이브러리
```

```python
1. 학습 방식에 따른 분류

 1) 지도 학습 (Supervised Learning)
    - 정답(라벨)이 있는 데이터를 학습
    - 입력 → 출력(정답)을 매칭하는 함수 학습
    - 대표 알고리즘: 선형 회귀, 로지스틱 회귀, 의사결정트리, 랜덤 포레스트, SVM, K-최근접 이웃(KNN), 신경망(딥러닝)
    - 예: 이메일 스팸 분류, 집값 예측  

  2) 비지도 학습 (Unsupervised Learning)
    - 정답이 없는 데이터에서 패턴, 군집, 구조를 찾는 방식
    - 대표 알고리즘: K-평균(K-means), 계층적 군집화, PCA(차원 축소), t-SNE
    - 예: 고객 세분화, 데이터 시각화

  3) 강화 학습 (Reinforcement Learning)
    - 에이전트(Agent)가 환경과 상호작용하며 보상(Reward)을 최대화하도록 학습
    - 대표 알고리즘: Q-Learning, Deep Q-Network(DQN), 정책 경사법, Actor-Critic
    - 예: 알파고, 자율 주행, 게임 AI
```

```python
2. 문제 유형에 따른 분류

 1) 회귀 (Regression) – 숫자 예측
   - 예: 주가 예측, 온도 예측

 2) 분류 (Classification) – 카테고리 예측
   - 예: 스팸 메일 여부, 암 진단(양성/음성)
   
 3) 클러스터링 (Clustering) – 비슷한 데이터끼리 묶기
   - 예: 고객 군집 분석

 4) 차원 축소 (Dimensionality Reduction) – 변수(특징) 축소
   - 예: 이미지 압축, 노이즈 제거
  
 5) 추천 시스템 (Recommendation)
   - 협업 필터링, 콘텐츠 기반 추천
   - 예: 넷플릭스 영화 추천, 쇼핑몰 상품 추천

 6) 이상 탐지 (Anomaly Detection)
    - 정상 패턴에서 벗어난 데이터를 탐지
    - 예: 금융 사기 탐지, 서버 장애 감지
```

```python
3. 모델 복잡도에 따른 분류

 1) 전통적 머신러닝
   - 회귀, 결정트리, 랜덤포레스트, SVM, KNN 등
   - 비교적 적은 데이터에서도 잘 동작

  2) 딥러닝 (Deep Learning)
   - 인공 신경망(ANN), CNN(이미지), RNN/LSTM(시계열), Transformer/BERT(자연어처리)
   - 대규모 데이터와 GPU 필요
```

```python
4. 응용 분야별 분류

 1) 컴퓨터 비전 → 이미지 분류, 얼굴 인식, 객체 탐지 (CNN, YOLO 등)
 2) 자연어 처리 (NLP) → 감성 분석, 번역, 챗봇 (Word2Vec, BERT, GPT 등)
 3) 시계열 데이터 → 주가 예측, 날씨 예측 (RNN, LSTM)
 4) 추천 시스템 → 넷플릭스, 아마존 추천
```

```python
1. 학습 방식에 따른 분류

## 1. 머신러닝의 큰 분류

머신러닝은 크게 다음과 같이 나눌 수 있습니다:

### (1) **지도 학습 (Supervised Learning)**

- 입력(X)과 정답(Y)이 있는 데이터를 학습
- 목표: X → Y를 잘 예측하는 모델 만들기
- 예시:
    - 스팸 메일 분류 (이메일 → 스팸/정상)
    - 집 값 예측 (특징 → 가격)
- 대표 알고리즘: **선형 회귀, 로지스틱 회귀, 결정 트리, 랜덤 포레스트, SVM, 신경망**

### (2) **비지도 학습 (Unsupervised Learning)**

- 정답이 없는 데이터에서  패턴, 군집, 구조를 찾는 방식
- 목표: 데이터 구조 이해, 그룹화, 차원 축소
- 예시:
    - 고객을 그룹별로 세분화 (클러스터링)
    - 데이터 시각화를 위한 차원 축소
- 대표 알고리즘: **K-Means, PCA,계층적 군집화, Autoencoder**

### (3) **강화 학습 (Reinforcement Learning)**

- 에이전트가 환경과 상호작용하며 보상을 최대화하는 방향으로 학습
- 예시:
    - 알파고(바둑)
    - 로봇 제어
- 개념: 상태(State), 행동(Action), 보상(Reward)

### (4) **딥러닝 (Deep Learning)**

- 신경망(Neural Network) 기반 학습
- 지도/비지도/강화학습 모두에 적용 가능
- 예시:
    - 이미지 분류(CNN)
    - 번역, 챗봇(RNN, LSTM, Transformer)
    - 음성 인식
```
