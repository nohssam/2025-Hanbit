## 검증 세트

#### 검증 세트 
> - 테스트 세트로 일반화 성능을 올바르게 예측하려면 가능한 한 테스트 세트를 사용하지 말아야 한다.<br>
    모델을 만들고 나서 마지막에 딱 한번만 사용하는 것이 좋다.
> - 테스트 세트를 사용하지 않고 이를 측정하는 간단한 방법은 훈련 세트를 또 나눠서 **검증세트**를 만드는 것이다.

```python
from IPython.display import Image, display
display(Image(filename="img/img61.png", width=800, height=600))
```

```python
import pandas as pd

wine = pd.read_csv('https://bit.ly/wine_csv_data')
data = wine[['alcohol', 'sugar', 'pH']]
target = wine['class']

data.shape
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)

print(train_input.shape, test_input.shape)
# 6497 * 0.8 = 5197.6,  6497 * 0.2 = 1,299.4
```

```python
sub_input, val_input, sub_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2, random_state=42)
# 5197 * 0.8 = 4157.6,  5197 * 0.2 = 1,039.4
print(sub_input.shape, val_input.shape)
```

```python
## 분류용 결정 트리 모델
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(sub_input, sub_target)

print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
```

#### 교차검증 이란 검증 세트를 떼어 내어 여러번 평가하는 과정를 말함

```python
display(Image(filename="img/img62.png", width=800, height=600))
# 아래 그림은 3-폴드 교차 검증이다.
# 보통 5-폴드 교차 검증이나 10-폴드 교차 검즈을 많이 사용한다.
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(dt, train_input, train_target)
print(scores)
# 항목	       의미	       설명
# fit_time	  훈련 시간	  각 fold(폴드=조각)별 모델 학습에 걸린 시간(초 단위)
# score_time  평가 시간    각 fold에서 테스트 데이터 평가에 걸린 시간
# test_score  검증 점수    각 fold별 검증 세트(테스트 부분)의 정확도
# 배열 크기가 5개인 이유는 기본 값인 5폴드를 사
```

```python
# 평균 정확도 
import numpy as np

print(np.mean(scores['test_score']))
```

```python
display(Image(filename="img/img63.png", width=800, height=600))
```

```python
# 분할기를 활용한 사용한 교차 검증
# StratifiedKFold : 분류 모델일 경우 타깃 클래스를 골고루 나누기(스트레트파이 K폴드 사용)
from sklearn.model_selection import StratifiedKFold
# 초기값은 5일때는 매갭녀수 생략 가능 
splitter = StratifiedKFold()
scores = cross_validate(dt, train_input, train_target, cv=splitter)
print(np.mean(scores['test_score']))
```

```python
# 10-폴드 교차 검증을 수행 했을 때 코딩이다.
splitter = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_validate(dt, train_input, train_target, cv=splitter)
print(np.mean(scores['test_score']))
```

## 그리드 서치

> - 여러 하이퍼 파라미터 조합을 **격자 형태로 전부 시도해보고 가장 성능이 좋은 조합을 찾아주는 방법**

### 하이퍼 파라미터 튜닝

```python
## 그리드 서치를 하는 이유 : 모델이 가장 잘 작동하는 설정(하이퍼파라미터)을 자동으로 찾아내기 위해서
# GridSearchCV가 자동으로 모든 조합을 시도하고 **가장 좋은 조합(best_params_)**을 골라줍니다.
# Grid(격자)는 여러 하이퍼파라미터 값을 모두 조합해서 탐색한다는 뜻
from sklearn.model_selection import GridSearchCV

# max_depth = None 일때 트리 깊이를 제한하지 않음
# 즉, 노드가 완전히 순수해질 때까지 계속 분할
#  0.0001 => 트리가 가지를 칠지 말지 결정하는 기준값(불순도 감소량)
# 불순도 감소량(줄어든 정도)이 0.0001보다 작으면 더 이상 가지치기를 하지 않는다”는 뜻
params = {'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004, 0.0005]}
gs = GridSearchCV(DecisionTreeClassifier(random_state=42), params, n_jobs=-1)
gs.fit(train_input, train_target)

# 가장 높은 CV 성능을 낸 하이퍼파라미터 조합
# GridSearchCV에서 가장 좋은 성능을 낸 최적의 모델 객체
dt = gs.best_estimator_

# 최적 모델(dt) 의 훈련 세트 정확도
print("훈련 세트의 정확도 :", dt.score(train_input, train_target))
print("테스트 세트의 정확도 :", dt.score(test_input,  test_target))
print('-' * 50)

# 가장 좋았던 하이퍼파라미터 조합(dict)
print("최적의 파라미터 : " ,gs.best_params_)
# 최적 교차검증 평균 점수
print("최적 교차검증 점수:", gs.best_score_)
print('-' * 50)

# 그리드에 넣은 모든 파라미터 조합에 대해, 각 조합의 교차검증 평균 점수 배열
print("모든 조합의 교차검증 평균 점수 : ", gs.cv_results_['mean_test_score'])
print("가장 높은 평균 교차검증 점수 : ", np.max(gs.cv_results_['mean_test_score']))

# cv_results_['params'](모든 파라미터 조합 리스트) 중에서, **최적 조합의 인덱스(best_index_)**에 해당하는 딕셔너리를 출력
# gs.best_params_와 동일
print(gs.cv_results_['params'][gs.best_index_])
```

```python
# GridSearchCV는 “모든 조합을 자동으로 시험” 해주는 도구
# “이 다섯 가지 값 중 어떤 게 모델 성능이 제일 좋은지 직접 실험해 봐!”  라고 시키는 거예요.

# max_depth 구하기
# from sklearn.model_selection import GridSearchCV

# params = {'max_depth': range(1, 20)}
# dt = DecisionTreeClassifier(random_state=42)
# gs = GridSearchCV(dt, params, cv=10)
# gs.fit(train_input, train_target)

# print("최적 하이퍼파라미터:", gs.best_params_)
# print("최적 교차검증 점수:", gs.best_score_)
```

```python
# min_impurity_decrease 구하기
# from sklearn.model_selection import GridSearchCV
# from sklearn.tree import DecisionTreeClassifier

# params = {'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004, 0.0005]}
# dt = DecisionTreeClassifier(random_state=42)

# gs = GridSearchCV(dt, params, cv=10)
# gs.fit(train_input, train_target)

# print("최적 하이퍼파라미터:", gs.best_params_)
# print("최적 교차검증 점수:", gs.best_score_)
```

```python
# # max_depth , min_impurity_decrease 같이 구하기
params = {
    'max_depth': range(1, 20),                         # ← max_depth 함께 탐색
    'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004, 0.0005]
}

# 재현성 있는 교차검증 설정(계층화 + 셔플 + 시드)
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

gs = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    params,
    cv=cv,
    n_jobs=-1,                 # 병렬
    return_train_score=True    # 과적합 점검용(선택)
)
gs.fit(train_input, train_target)

# 가장 높은 CV 성능을 낸 하이퍼파라미터 조합
# GridSearchCV에서 가장 좋은 성능을 낸 최적의 모델 객체
dt = gs.best_estimator_

# 최적 모델(dt)의 훈련/테스트 세트 정확도
print("훈련 세트의 정확도 :", dt.score(train_input, train_target))
print("테스트 세트의 정확도 :", dt.score(test_input,  test_target))
print('-' * 50)

# 최적 조합 및 CV 성능
print("최적의 파라미터 :", gs.best_params_)        # 예: {'max_depth': 7, 'min_impurity_decrease': 0.0002}
print("최적 교차검증 점수:", gs.best_score_)
print('-' * 50)

# 그리드의 모든 조합 결과 요약
print("모든 조합의 교차검증 평균 점수 :", gs.cv_results_['mean_test_score'])
print("가장 높은 평균 교차검증 점수 :", np.max(gs.cv_results_['mean_test_score']))
print('-' * 50)
# 최적 조합(= best_params_) 재확인
print("best_index_로 확인한 최적 조합 :", gs.cv_results_['params'][gs.best_index_])
print('-' * 50)
# (선택) 과적합 여부 빠르게 확인: train vs test CV 평균
# 조합별 평균 train/test 점수를 함께 보고 싶다면 아래 두 줄도 참고
# print(gs.cv_results_['mean_train_score'])
# print(gs.cv_results_['mean_test_score'])
```

## 랜덤 서치
> - GridSearch보다 훨씬 효율적으로 하이퍼파라미터를 탐색하기 위해서
> - GridSearch처럼 모든 조합을 다 보는 게 아니라, 일정 횟수만 “무작위로” 조합을 골라서 탐색

```python
from scipy.stats import uniform, randint
# 0부터 10 직전까지(즉 [0, 10))의 정수에서 균등하게 뽑는 분포 객체.
rgen = randint(0, 10)
# 정수 10개 무작위 추출. (rvs = random variates)
rgen.rvs(10)
```

```python
# 정수 1000개 뽑아 각 값이 몇 번 나왔는지 집계.
np.unique(rgen.rvs(1000), return_counts=True)
```

```python
# 구간 [loc, loc+scale) = [0, 1) 에서 실수를 균등 추출.
# 실수 10개를 균등하게 추출
ugen = uniform(0, 1)
ugen.rvs(10)
```

```python
params = {'min_impurity_decrease': uniform(0.0001, 0.001),
          'max_depth': randint(20, 50),
          'min_samples_split': randint(2, 25),
          'min_samples_leaf': randint(1, 25),
          }
```

```python
from sklearn.model_selection import RandomizedSearchCV

rs = RandomizedSearchCV(DecisionTreeClassifier(random_state=42), params,
                        n_iter=100, n_jobs=-1, random_state=42)
rs.fit(train_input, train_target)
print(rs.best_params_)
print(np.max(rs.cv_results_['mean_test_score']))

dt = rs.best_estimator_
print(dt.score(test_input, test_target))
```

```python
from sklearn.model_selection import RandomizedSearchCV

# 탐색할 범위 지정
params = {
    'max_depth': np.arange(1, 20),
    'min_impurity_decrease': np.linspace(0.0001, 0.001, 10)
}

rs = RandomizedSearchCV(
    DecisionTreeClassifier(random_state=42),
    params,
    n_iter=10,       # 10개의 무작위 조합만 시도
    cv=5,
    random_state=42,
    n_jobs=-1
)
rs.fit(train_input, train_target)

print("최적 파라미터:", rs.best_params_)
print("최적 교차검증 점수:", rs.best_score_)
print('*' * 50)
best_dt = rs.best_estimator_
print("훈련 세트 정확도:", best_dt.score(train_input, train_target))
print("테스트 세트 정확도:", best_dt.score(test_input, test_target))
```
