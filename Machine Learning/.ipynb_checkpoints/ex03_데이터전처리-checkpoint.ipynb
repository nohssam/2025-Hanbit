## 데이터 전처리

```python
from IPython.display import Image, display
display(Image(filename="img/img18.png", width=200, height=200))

# 길이가 25cm , 150g 인 도미인데 앞에서의 모델은 빙어라고 예측
```

```python
import numpy as np

fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

fish_data = [[l, w] for l, w in zip(fish_length, fish_weight)]
fish_target = [1]*35 + [0]*14

from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()

input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

np.random.seed(42)
index = np.arange(49)
np.random.shuffle(index)

# 훈련 셋트
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]

# 테스트 셋트
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]

kn.fit(train_input, train_target)

new = [[25,150]]

kn.predict(new)  

# 결과 : array([0]) => 0 이므로 빙어로 나옴 
```

```python
display(Image(filename="img/img19.png", width=600, height=600))
```

```python
# 리스트 \ 방향으로 합치기
np.concatenate(([1,2,3], [4,5,6]))
```

```python
fish_data = np.column_stack((fish_length, fish_weight))
#print(fish_data)
```

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
# print(fish_target)
```

## 사이킷런으로 훈련 세트와 테스트 세트 나누기

```python
display(Image(filename="img/img20.png", width=600, height=600))
```

```python
from sklearn.model_selection import train_test_split
```

```python
# random_state=42 의 의미는 데이터를 무작위로 섞을 때 사용할 난수 발생기의 시드(seed) 
# stratify=fish_target => 데이터 분할 시 클래스 비율(라벨 비율)을 유지하도록 (샘플링 평향 방지)
# (훈련셋트가 도미와 빙어가 2:1 이면 테스트셋트도 도미와 빙어가 2:1 로 유지 )
train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, stratify=fish_target ,random_state=42)

from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
# 수상한 객체 판별하기 
print(kn.predict([[25, 150]]))
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 주변에 있는 샘플의 주변 샘플의 가장 가까운 찾아 준다. => distances
distances, indexes = kn.kneighbors([[25, 150]])
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(train_input[indexes])
```

```python
print(train_target[indexes])
```

```python
# 5개의 거리 
print(distances)
```

## 기준을 맞춰라 

### 스케일에 영향을 맞는다.
##### - k-최근, 선형회기, 로지스티 회기 
##### - 딥러닝 부분 - 인공신경망 알고리즘들

```python
# x축과 y축의 단위가 달라서 그렇다.
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlim((0, 1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 표준점수로 바꾸기 : (평균 - 값) / 표준편차

```python
# 평균과 표준편차 구하기 (axis=0 은 열 기준이다.)
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
print(mean, std)
```

```python
# (평균 - 값) / 표준편차
train_scaled = (train_input - mean) / std
```

```python
display(Image(filename="img/img21.png", width=600, height=600))
```

### 전처리 데이터로 모델 훈련하기

```python
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
new = ([25, 150] - mean) / std
```

```python
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.fit(train_scaled, train_target)
```

```python
test_scaled = (test_input - mean) / std
```

```python
kn.score(test_scaled, test_target)
```

```python
print(kn.predict([new]))
```

```python
distances, indexes = kn.kneighbors([new])
```

```python
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes,0], train_scaled[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
