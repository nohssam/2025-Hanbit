### 경사 하강법

#### “확률적 경사 하강법 (Stochastic Gradient Descent, SGD)”은 딥러닝과 머신러닝에서 가장 널리 쓰이는 최적화 알고리즘이에요.

기본 개념: “기울기를 따라 내려가기”
**경사 하강법 (Gradient Descent)** 은 말 그대로 **오차(loss)를 최소화하기 위해 기울기를 따라 내려가는 방법**
> 경사 하강법은 "기울기(gradient)" 방향으로 조금씩 내려가면서 최소점을 찾는 과정

##### 일반적인 배치 경사 하강법 (Batch Gradient Descent)
> - 모든 데이터를 한 번에 사용합니다. <br>
> - 한 번의 업데이트에 전체 데이터를 다 써서 계산량이 큼

##### 확률적 경사 하강법 (Stochastic Gradient Descent, SGD)
> **데이터 한 개(또는 아주 작은 일부)**를 사용 <br>
> 매번 임의로 하나의 샘플을 골라서 기울기를 계산하고 즉시 업데이트
> 빠르지만 노이즈(잡음)가 많음 → 오차 곡선을 따라 “덜컹덜컹” 내려감

##### 미니배치 경사 하강법 (Stochastic Gradient Descent, SGD)
> - 실제로 가장 많이 쓰이는 방식입니다.
> - 전체 데이터 중 일부(m개) 샘플만 뽑아서 평균 기울기 계산
> - 효율성과 안정성의 절충안
> - 대부분의 딥러닝 프레임워크 (TensorFlow, PyTorch)는 기본적으로 이 방식 사용

### 경사 하강법을 “산 내려가기”로 생각하면 이해가 쉽습니다.
> - Batch GD: 산 전체를 다 살펴보고 가장 좋은 방향으로 한 걸음 <br>
> - SGD: 눈앞의 한 지점만 보고 내려감 (빠르지만 불안정)<br>
> - Mini-batch: 여러 점을 평균 내서 방향을 결정 (가장 현실적)

### 로지스틱 손실함수

이런 정확도는 어떤 임의의 실수로 출력이 되지 않는다. <br>
그래서 손실함수를 사용하기 어렵다. <br>
정확도를 분류문제에 손실함수로 사용할 수 없다. <br>
**측정용도로는 모델의 정확한지 모델의 성능이 좋은지는 정확도로 예측**할 수 있지만  <br>
**손실 함수 경사하강법을 사용하기 위한 손실 함수로는 사용하기 어렵다.**  <br>
정확도를 올리수 있는 유사한 손실함수를 대리로 만든다. 이것을 **대리 손실 함수**라고 한다.

로그 손실(Log Loss) 또는 이진 크로스엔트로피(Binary Cross Entropy) 의 원리를 설명하는 문장

- 양성 클래스 (타킷 = 1) 일때 손실은 -log(예측확률)계산 <br>
    => 확률이 1에서 멀어져 0에 가까워질 수록 손실은 아주 큰 양수가 된다. <br>
- 음성 클래스 (타킷 = 0) 일때 손실은 -log(1-예측확률)계산 <br>
    => 확률이 0에서 멀어져 1에 가까워질 수록 손실은 아주 큰 양수가 된다.

### 확률적 경사 하강법을 사용한 분류 모델 만들어보자 

```python
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish_input = fish[['Weight','Length','Diagonal','Height','Width']]
fish_target = fish['Species']
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
```

```python
# 훈련 세트와 테스트 세트의 특성을 표준화 전처리
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# 사이킷 런에서 확률적 경사 하강법을 사용하기 위해서 SGDClassifier 클래스사용
from sklearn.linear_model import SGDClassifier
```

```python
# loss='log_loss'는 손실함수 지정(즉, 로지스틱 손실함수지정), 
#  max_iter=10는 에포크 횟수 지정 (전체 훈련 세트 10회 반복), 학습량이 작어서 오류 발생
display(Image(filename="img/img50.png", width=600, height=400))
display(Image(filename="img/img51.png", width=600, height=400))

sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)

# 훈련, 테스트 세트 정확도 출력
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
sc.partial_fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

### 에포크와 과대/과소적합

```python
display(Image(filename="img/img52.png", width=600, height=400))
```

```python
import numpy as np

sc = SGDClassifier(loss='log_loss', random_state=42)

train_score = []
test_score = []

classes = np.unique(train_target)

for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target, classes=classes)

    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

### 반복횟수를 100으로 수정해서 다시 

```python
sc = SGDClassifier(loss='log_loss', max_iter=100,  random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
# tol은 “손실 감소가 이 정도 이하로 작으면 학습을 멈춰라”는 기준입니다.
# 너무 작게 잡으면 계속 학습하게 되고, 너무 크게 잡으면 조기 종료될 수도 있어요.
sc = SGDClassifier(loss='log_loss', max_iter=100, tol=None,random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```
