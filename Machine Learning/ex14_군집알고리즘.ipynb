## 군집알고리즘
>- 비지도 학습(Unsupervised Learning) 의 한 종류
>- 정답(레이블) 이 없는 데이터를 유사한 특징끼리 묶는(그룹화하는) 방법
#### 군집의 목적
>- 데이터의 숨겨진 패턴이나 구조를 발견
>- 고객을 성향별로 나누는 시장 세분화(Market Segmentation)
>- 이미지, 문서, 유전자 데이터 등에서 유사성 기반 분류
>- 이상치(Outlier) 탐지

#### 군집 알고리즘의 종류

#### (1) K-평균 군집(K-Means Clustering)
>- 가장 기본적이고 많이 사용하는 방법
>- 군집의 개수 K 를 미리 지정해야 함
>- 중심점(centroid)을 기준으로 데이터를 반복적으로 재배치
#### 작동 원리 
>- K개의 임의 중심점을 설정
>- 각 데이터를 가장 가까운 중심점에 배정
>- 각 군집의 평균 위치로 중심점을 갱신
>- 중심점이 더 이상 변하지 않을 때까지 반복

```python
# 3종류의 과일 이미지를 ‘평평하게(flatten)’ 만들어서,
#  머신러닝 모델에 넣기 좋은 형태로 바꾸는 작업
# fruits[0:100]  0 ~ 99  사과 100장
# reshape(-1, 100*100) =>  reshape은 배열 모양을 바꾸는 함수
# 원래: (100, 100, 100) → 100장의 100×100 이미지
# 변환 후: (100, 10000) → 100장의 1차원 벡터 (100×100 = 10,000픽셀)
# 각 이미지를 평평하게(Flatten) 만들어서 하나의 행(row)에 모든 픽셀값(10,000개)을 나열한 형태로 만듭니다.
apple = fruits[0:100].reshape(-1, 100*100)
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)
```

```python
apple_mean = apple.mean(axis=0).reshape(100, 100)
pineapple_mean = pineapple.mean(axis=0).reshape(100, 100)
banana_mean = banana.mean(axis=0).reshape(100, 100)

fig, axs = plt.subplots(1, 3, figsize=(20, 5))
axs[0].imshow(apple_mean, cmap='gray_r')
axs[1].imshow(pineapple_mean, cmap='gray_r')
axs[2].imshow(banana_mean, cmap='gray_r')
plt.show()
```

### 평균값 과 가까운 사진 고르기

```python
# fruits : (300, 100, 100) - apple_mean : (100, 100)
# 모든 이미지(사과/파인애플/바나나 각각)에 대해 “사과 평균 이미지”와의 픽셀 차이(밝기 차이) 계산
# 음수 없이 “얼마나 다르냐”만 남김
abs_diff = np.abs(fruits - apple_mean)
# 이제 각 이미지마다, 100×100개의 픽셀 차이를 평균냅니다.
# 한 이미지 전체가 “사과 평균 이미지”와 얼마나 다른지(평균 차이값)를 한 숫자로 요약
abs_mean = np.mean(abs_diff, axis=(1,2))
# 300개의 값이 생김 → 300장의 이미지 각각에 대한 “사과와의 평균 차이 정도”
# abs_mean 값이 작으면 → 그 이미지는 사과 평균과 비슷한 이미지
# abs_mean 값이 크면 → 그 이미지는 사과 평균과 다른 이미지
print(abs_mean.shape)
```

```python
# 사과 평균 이미지와 가장 비슷한 과일들”을 실제로 화면에 보여주는 시각화 코드
# abs_mean을 오름차순(작은 값부터) 정렬했을 때의 인덱스 번호를 반환
# 사과 평균과 가장 비슷한 이미지부터 가장 다른 이미지까지 순서를 알려줍니다
# 가장 가까운(비슷한) 100장의 이미지의 인덱스만
apple_index = np.argsort(abs_mean)[:100]
# apple_index의 원래 형태는 (100,)
# 이걸 reshape(10,10) 으로 바꿔서  10×10 격자 모양으로 이미지를 표시하기 쉽게 만듭니다
apple_index = apple_index.reshape(10, 10)

fig, axs = plt.subplots(10, 10, figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[apple_index[i, j]], cmap='gray_r')
        axs[i, j].axis('off')
plt.show()
```

>- 픽셀값을 사용해 과일 사진을 모으는 작업을 해 보았습니다.
>- 이렇게 비슷한 샘플끼리 그룹으로 모으는 작업을 **군집** 이라고 합니다.
>- **군집**은 대표적인 **비지도학습 작업 중 하나** 이다. 
>- 군집알고리즘에서 만든 그룹을 **클러스터**라고 한다.

##### “바나나 평균 이미지(banana_mean)와의 평균 절댓값 차이가 작은 순서”로 100장을 뽑다 보니, 끝부분(경계) 에서 사과 몇 장이 끼어든 겁니다.
##### 왜 사과가 섞였을까?
- 1.지표의 한계
>- 우리는 mean(|image - banana_mean|)(픽셀 단위 L1 평균)만 봤어요.
>- 해상도·조명·그림자·회전 때문에 몇몇 사과가 바나나 평균과 픽셀 통계가 비슷해질 수 있어요(특히 둥근 하이라이트 + 어두운 가장자리 패턴).
- 2.클래스 간 분포 겹침
>- 바나나 100장의 거리는 분포가 있고, 사과 100장도 거리가 분포가 있어요.
>- 두 분포의 꼬리(상위/하위 퍼센타일)가 겹치면 “최상위 100장”의 끝부분에서 타 클래스가 끼어듭니다.
>- 그림의 마지막 줄에 사과가 보이는 건 바로 이 **경계부 샘플**이란 뜻.
- 3.단일 프로토타입(평균) 기준
>-  바나나는 포즈·곡률이 다양해서 하나의 평균 이미지로는 모든 바나나를 잘 대표하기 어렵습니다.
>- 이럴수록 타 클래스 일부가 평균에 더 가깝게 측정될 수 있어요.

```python
abs_diff = np.abs(fruits - banana_mean)
abs_mean = np.mean(abs_diff, axis=(1,2))

banana_index = np.argsort(abs_mean)[:100]
banana_index = banana_index.reshape(10, 10)
fig, axs = plt.subplots(10, 10, figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[banana_index[i, j]], cmap='gray_r')
        axs[i, j].axis('off')
plt.show()
```

```python
abs_diff = np.abs(fruits - pineapple_mean)
abs_mean = np.mean(abs_diff, axis=(1,2))

pineapple_index = np.argsort(abs_mean)[:100]
pineapple_index = pineapple_index.reshape(10, 10)
fig, axs = plt.subplots(10, 10, figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[pineapple_index[i, j]], cmap='gray_r')
        axs[i, j].axis('off')
plt.show()
```
