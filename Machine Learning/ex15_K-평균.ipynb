```python
display(Image(filename="img/img69-2.jpg", width=800, height=600))
# 1. 중심점 선택 : 랜덤하게 초기 중심점을 선택한다.
# 2. 클러스터 할당 : k개의 중심점과 각각의 개별 데이터간의 거리를 측정 후,
#                   가장 가까운 중심점을 기준으로 데이털르 할당한다.
#                   이 과정에을 통해 클러스터가 구성된다.
# 3. 새로운 중심점 선택 : 클러스터마다 새로운 중심점을 계산한다.
# 4. 범위 확인 : 선택된 중심점에 더 이상의 변화가 없다면 진행을 멈춘다.
#                만약 계속 변화가 있다면 반복한다.
```

```python
import numpy as np
fruits = np.load('data/fruits_300.npy')
# reshape()은 배열의 모양을 바꾸는 함수입니다.
# 3차원 이미지를 2차원 표로 변환
# -1은 자동 계산(auto)
# 이유 : 이미지를 평평하게(flatten) 바꾸면, 각 이미지를 하나의 "데이터 행(row)"로 만들어 머신러닝 모델에 넣기 쉽다.
fruits_2d = fruits.reshape(-1, 100*100)
```

```python
# 사이킷런(sklearn)의 군집 모듈(cluster) 에서 K-평균 알고리즘을 불러
from sklearn.cluster import KMeans
# n_clusters=3, 만들 군집(그룹) 개수 → 과일 종류가 3가지이므로 3으로 설정
km = KMeans(n_clusters=3, random_state=42)
# 훈련시키자
km.fit(fruits_2d)
```

```python
# km.labels_	각 데이터(이미지)가 속한 군집 번호	[0, 0, 1, 2, 1, ...]
# km.cluster_centers_	3개 군집의 중심(평균) 위치	(3, 10000) 크기의 배열
# km.inertia_	군집 내 거리(오차)의 총합	값이 작을수록 잘 뭉쳐 있음
print(km.labels_)
print('*' * 50)
print(km.cluster_centers_)
print('*' * 50)
print(km.inertia_)
print('*' * 50)
```

```python
# 군집을 중심으로 이미지 보
import matplotlib.pyplot as plt

for label in range(3):
    center = km.cluster_centers_[label].reshape(100, 100)
    plt.subplot(1, 3, label + 1)
    plt.imshow(center, cmap='gray_r')
    plt.title(f'Cluster {label}')
    plt.axis('off')
plt.show()
```

```python
# 레이블값 0,1,2이다. 순서에는 어떤 의미도 없음 
# 레이블의 샘플의 개수이다.
print(np.unique(km.labels_, return_counts=True))
```

```python
# 여러 장의 과일 이미지를 한눈에 보기 좋게 격자(grid) 형태로 그려주는 도우미 함수
# arr : 이미지 배열, ratio : 이미지 크기 비율
def draw_fruits(arr, ratio=1):
    n = len(arr)    # n은 샘플 개수입니다
    # 한 줄에 10개씩 이미지를 그립니다. 샘플 개수를 10으로 나누어 전체 행 개수를 계산합니다.
    rows = int(np.ceil(n/10))
    # 행이 1개 이면 열 개수는 샘플 개수입니다. 그렇지 않으면 10개입니다.
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols,
                            figsize=(cols*ratio, rows*ratio), squeeze=False)
    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:    # n 개까지만 그립니다.
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()
```

```python
draw_fruits(fruits[km.labels_==0])
```

```python
draw_fruits(fruits[km.labels_==1])
```

```python
draw_fruits(fruits[km.labels_==2])
```

### 클러스터 중심
> - cluster_centers_ 속성에 클러스터 중심이 저장되어 있음
> - 이전에 사용한 것과 비슷하다

```python
draw_fruits(km.cluster_centers_.reshape(-1, 100, 100), ratio=3)
```

##### K평균 클래스는 훈련 데이터 샘플에서 클러스터 중심까지 거리로 변환해 주는  transform()메서드를 가지고 있음
##### StandardScaler 클래스 처럼 특성값을 변환하는 도구로 사용할 수 있다.

```python
# 새로운 데이터(여기선 101번째 과일)가 각 군집 중심과 얼마나 떨어져 있는지 거리값으로 출력
# transform → “얼마나 떨어졌나?”
# predict → “어디에 속하나?”
# draw_fruits → “실제 이미지 보기”
# n_iter_ → “학습 반복 횟수 확인”

print(km.transform(fruits_2d[100:101]))
# 3400.24197319 : 사과 중심까지 거리
# 8837.37750892 : 파인애플 중심까지 거리
# 5279.33763699 : 바나나 중심까지 거리
```

```python
# 101번째 과일은 어느 군집(cluster)으로 분류되는가?
print(km.predict(fruits_2d[100:101]))
```

```python
draw_fruits(fruits[100:101])
```

```python
# K-Means가 학습 중 중심을 몇 번 갱신(반복)했는가
print(km.n_iter_)
```

### 최적의 K (클러스터) 찾기

```python
# 위 코딩을 할때 우리는 클러스터 개수를 3개라고 지정했다.
# 원래는 최적의 k(클러스터의 개수)를 찾아야 한다
# 2-6까기 바꿔가며 k평균클래스를 5번 훈련한 결과
inertia = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(fruits_2d)
    inertia.append(km.inertia_)

plt.plot(range(2, 7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```
